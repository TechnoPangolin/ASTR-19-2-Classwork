```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
setRand = np.random.gumbel(0.5, 1.0,1000)

f, ax, = plt.subplots(1,1)

ax.hist(setRand, facecolor = 'red', bins=100, alpha=0.5)
ax.set_xlabel("x")
ax.set_ylabel("values of x")

plt.tight_layout()
plt.savefig("CJ2_p10 plot 1.pdf", dpi=400, bbox_inches='tight', format='pdf')
```
