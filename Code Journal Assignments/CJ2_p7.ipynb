import required libraries as their aliases for ease of use

```python
import matplotlib.pyplot as plt
import numpy as np
```

Initialize x with the required values

```python
x = np.linspace(0, 1, 100)
```

define the function that will return an array where each value of the input array is squared

```python
def exp(x):
    return(np.exp(x))
```

define y as each value of x squared, and graph the two arrays per the requirements

```python
y = exp(x)

f, ax = plt.subplots(1, 1)

ax.plot(x, y, color='red', linewidth=1.5)

ax.set_xlabel("Time (milliseconds)")
ax.set_ylabel("Awesomeness")

plt.savefig("CJ2_p7 plot 1.pdf", bbox_inches='tight', dpi=400, format='pdf')
```
