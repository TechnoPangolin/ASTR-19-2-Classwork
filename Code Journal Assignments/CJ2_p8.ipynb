import the required libraries as their aliases

```python
import numpy as np
import matplotlib.pyplot as plt
```

initialize x with the required range, define the two required sin and cos functions

```python
x = np.linspace(0, 1, 100)

def sin(x):
    return np.sin(x)

def cos(x):
    return np.cos(x)
```

generate a multi plot layout for the two sets of values, fix its formatting, and save it.

```python
f, ax = plt.subplots(1,2)

ax[0].plot(x, sin(x))
ax[0].set_xlabel("x")
ax[0].set_ylabel("sin(x)")

ax[1].plot(x, cos(x))
ax[1].set_xlabel("x")
ax[1].set_ylabel("cos(x)")

plt.tight_layout()
plt.savefig("CJ2_p8 plot 1.pdf", bbox_inches='tight', dpi=400, format='pdf')
```
