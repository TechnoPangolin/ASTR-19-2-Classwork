```python
import numpy as np
import matplotlib.pyplot as plt
```

initialize a set of random numbers meeting the requirements, then plot them on a histogram with 100 bins of separation.

```python
setRand = np.random.rand(1000)

f, ax, = plt.subplots(1,1)

ax.hist(setRand, facecolor = 'red', bins=100, alpha=0.5)
ax.set_xlabel("x")
ax.set_ylabel("values of x")

plt.tight_layout()
plt.savefig("CJ2_p9 plot 1.pdf", dpi=400, bbox_inches='tight', format='pdf')
```
